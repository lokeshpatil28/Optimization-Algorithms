# Optimization Methods Project

Implementation of Bisection, Golden Section, Fibonacci, and Steepest Ascent methods
with automatic numerical differentiation and graphical outputs.

```
import numpy as np
import math
import matplotlib.pyplot as plt
import os

from Bisection_method import run_bisection
from Golden_Section import run_golden
from Fiibonacci_method import run_fibonacci
from Steepest_Ascent import run_steepest
```

```
safe_dict = {
    "x": 0,
    "y": 0,
    "np": np,
    "math": math,
    "sin": math.sin,
    "cos": math.cos,
    "exp": math.exp,
    "log": math.log,
    "sqrt": math.sqrt
}

def plot_1d_function(f, x_opt, title):
    x_vals = np.linspace(x_opt - 5, x_opt + 5, 400)
    y_vals = [f(x) for x in x_vals]

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, y_vals)
    plt.axvline(x=x_opt, linestyle='--')
    plt.scatter([x_opt], [f(x_opt)])

    os.makedirs("outputs", exist_ok=True)
    filename = title.replace(" ", "_") + ".png"
    path = os.path.join("outputs", filename)
    plt.savefig(path)
    print(f"Saved in outputs/{filename}")

    plt.show()

def plot_2d_function(f, x_opt, y_opt):
    x = np.linspace(x_opt - 5, x_opt + 5, 100)
    y = np.linspace(y_opt - 5, y_opt + 5, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.vectorize(f)(X, Y)

    plt.contourf(X, Y, Z)
    plt.scatter(x_opt, y_opt)

    os.makedirs("outputs", exist_ok=True)
    path = os.path.join("outputs", "Steepest_Ascent_Result.png")
    plt.savefig(path)
    print("Saved in outputs/Steepest_Ascent_Result.png")

    plt.show()

def numerical_derivative(f, h=1e-5):
    return lambda x: (f(x+h) - f(x-h))/(2*h)

def numerical_gradient(f, h=1e-5):
    return lambda x,y: np.array([
        (f(x+h,y)-f(x-h,y))/(2*h),
        (f(x,y+h)-f(x,y-h))/(2*h)
    ])

def get_function_1d():
    expr = input("Enter function f(x): ")
    return lambda x: eval(expr, {"__builtins__": None}, {**safe_dict, "x": x})

def get_function_2d():
    expr = input("Enter function f(x,y): ")
    return lambda x,y: eval(expr, {"__builtins__": None}, {**safe_dict, "x": x, "y": y})
```

```
def main():
    print("1. Bisection\n2. Golden\n3. Fibonacci\n4. Steepest\n5. All")
    choice = int(input("Choice: "))

    if choice == 5:
        f_common = get_function_1d()

    if choice in [1,5]:
        f = f_common if choice==5 else get_function_1d()
        df = numerical_derivative(f)
        a = float(input("a: "))
        b = float(input("b: "))
        root,_ = run_bisection(df,a,b)
        print(root)
        plot_1d_function(f, root, "Bisection")

    if choice in [2,5]:
        f = f_common if choice==5 else get_function_1d()
        a = float(input("a: "))
        b = float(input("b: "))
        x,_ = run_golden(f,a,b)
        print(x)
        plot_1d_function(f, x, "Golden")

    if choice in [3,5]:
        f = f_common if choice==5 else get_function_1d()
        a = float(input("a: "))
        b = float(input("b: "))
        n = int(input("n: "))
        x,_ = run_fibonacci(f,a,b,n)
        print(x)
        plot_1d_function(f, x, "Fibonacci")

    if choice in [4,5]:
        f2d = get_function_2d()
        grad = numerical_gradient(f2d)
        x0 = float(input("x0: "))
        y0 = float(input("y0: "))
        alpha = float(input("alpha: "))
        x,y,_ = run_steepest(f2d,grad,x0,y0,alpha)
        print(x,y)
        plot_2d_function(f2d,x,y)

main()
```
